# An치lisis de intensidad emocional

Ejemplo minimalista que usa el paquete `prueba_emociones` para entrenar un modelo cl치sico sobre el dataset `stepp1/tweet_emotion_intensity`.

```python
from prueba_emociones import EmotionModel, load_hf_dataset

# Cargar un subconjunto para iterar r치pido
df, text_col, label_col = load_hf_dataset(
    "stepp1/tweet_emotion_intensity",
    split="train",
    text_column="tweet",
    label_column="intensity",
    limit=500,
)

df.head()
```

```python
# Entrenar modelo base
model = EmotionModel()
result = model.fit(df[text_col], df[label_col], test_size=0.2, random_state=42)
print(f"Accuracy: {result.accuracy:.3f}")
print(result.report)
```

```python
# Ejemplos de predicci칩n
ejemplos = [
    "This is okay",
    "I ABSOLUTELY LOVE THIS!!!",
    "Kinda disappointing...",
]

model.predict(ejemplos)
```
