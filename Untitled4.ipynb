```python
import jellyfish

def corregir_palabra(palabra, lista_objetivo, umbral=0.92):
    """
    Compara una palabra con una lista (intensificadores, atenuadores, negadores)
    usando Jaro-Winkler. Si es similar, devuelve la palabra correcta.
    """
    palabra = palabra.lower()
    mejor = palabra
    mejor_score = 0

    for w in lista_objetivo:
        score = jellyfish.jaro_winkler_similarity(palabra, w)
        if score > mejor_score:
            mejor_score = score
            mejor = w

    return mejor if mejor_score >= umbral else palabra


def corregir_texto(tweet, listas, umbral=0.92):
    """
    Corrige un tweet palabra por palabra, SOLO si coincide con tus diccionarios.
    """
    palabras = tweet.split()
    corregidas = []

    for p in palabras:
        original = p
        for categoria in listas.values():     # intensificadores, atenuadores, negadores
            for idioma, lista_palabras in categoria.items():
                p = corregir_palabra(p, lista_palabras, umbral)
        corregidas.append(p)

    return " ".join(corregidas)
```

```python
import pandas as pd
import numpy as np
import re
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek
from datasets import load_dataset
import warnings
warnings.filterwarnings('ignore')

# =====================================================
# SISTEMA DE LÓGICA DIFUSA MEJORADO (15 REGLAS)
# =====================================================

class SistemaLogicaDifusaMejorado:
    """
    Sistema de inferencia difusa con:
    - Características paralingüísticas
    - Reglas lingüísticas (atenuadores, intensificadores, negaciones)
    - 15 reglas difusas mejoradas
    """

    def __init__(self):
        # Diccionarios lingüísticos multiidioma
          self.intensificadores = {
              'en': [
                  'very', 'extremely', 'absolutely', 'totally', 'completely', 'really',
                  'so', 'too', 'super', 'incredibly', 'amazingly', 'utterly',
                  # Nuevos
                  'sooo', 'reallyyy', 'veryyy', 'soo', 'superrr',
                  'literally', 'insanely', 'crazy', 'hella', 'mad', 'mega', 'ultra',
                  'way too', 'so damn', 'fricking', 'freaking', 'damn'
              ],
              'es': [
                  'muy', 'extremadamente', 'absolutamente', 'totalmente', 'completamente',
                  'realmente', 'súper', 'demasiado', 'increíblemente', 'sumamente',
                  # Nuevos
                  'muyyy', 'superrr', 're', 'requete', 'ultra', 'mega',
                  'demasiaaado', 'demasiadoo', 'tan', 'taaan', 'pero muy',
                  'literal', 'literalmente', 'bien', 'recontra', 'hyper'
              ]
          }

          self.atenuadores = {
              'en': [
                  'somewhat', 'slightly', 'a bit', 'kind of', 'sort of', 'rather',
                  'pretty', 'fairly', 'quite', 'almost', 'barely', 'hardly',
                  # Nuevos
                  "kinda", "sorta", "a lil", "lil", "lowkey", "mildly",
                  "not really", "not quite", "just a bit"
              ],
              'es': [
                  'algo', 'un poco', 'medio', 'casi', 'apenas', 'ligeramente',
                  'más o menos', 'relativamente', 'bastante',
                  # Nuevos
                  'poquito', 'un tantito', 'medio que', 'no tanto', 'suave',
                  'apenas si', 'como que'
              ]
          }

          self.negadores = {
              'en': [
                  'not', 'no', 'never', 'nothing', 'nowhere', 'neither', 'nobody',
                  "n't", "dont", "doesnt", "wont", "cant", "shouldnt",
                  # Nuevos
                  "ain't", "nah", "nope", "never ever", "cant even",
                  "dont even", "doesnt even"
              ],
              'es': [
                  'no', 'nunca', 'nada', 'ningún', 'ninguno', 'jamás', 'tampoco',
                  # Nuevos
                  'naa', 'nel', 'nopo', 'para nada', 'ni de broma',
                  'ni en pedo', 'ni loco', 'jamássss'
              ]
          }
          self.listas_correccion = {
              "intensificadores": self.intensificadores,
              "atenuadores": self.atenuadores,
              "negadores": self.negadores
          }



    # ==================== FUNCIONES DE PERTENENCIA ====================

    def membresia_triangular(self, x, a, b, c):
        """Función de pertenencia triangular"""
        if x <= a or x >= c:
            return 0.0
        elif a < x <= b:
            return (x - a) / (b - a)
        else:
            return (c - x) / (c - b)

    def membresia_trapezoidal(self, x, a, b, c, d):
        """Función de pertenencia trapezoidal"""
        if x <= a or x >= d:
            return 0.0
        elif a < x <= b:
            return (x - a) / (b - a)
        elif b < x <= c:
            return 1.0
        else:
            return (d - x) / (d - c)

    # ==================== FUZZIFICACIÓN PARALINGÜÍSTICA ====================

    def fuzzificar_exclamaciones(self, n_exclamaciones):
        """Convierte número de exclamaciones a grados difusos"""
        return {
            'ninguna': max(0, 1 - n_exclamaciones),
            'pocas': self.membresia_triangular(n_exclamaciones, 0, 1, 3),
            'moderadas': self.membresia_triangular(n_exclamaciones, 1, 3, 6),
            'muchas': max(0, min(1, (n_exclamaciones - 3) / 5))
        }

    def fuzzificar_mayusculas(self, ratio_mayusculas):
        """Convierte ratio de mayúsculas a grados difusos"""
        return {
            'bajo': max(0, 1 - ratio_mayusculas * 10),
            'medio': self.membresia_triangular(ratio_mayusculas, 0.05, 0.2, 0.5),
            'alto': max(0, min(1, (ratio_mayusculas - 0.2) / 0.3))
        }

    def fuzzificar_repeticiones(self, n_repeticiones):
        """Convierte repeticiones de caracteres a grados difusos"""
        return {
            'bajo': max(0, 1 - n_repeticiones / 2),
            'medio': self.membresia_triangular(n_repeticiones, 1, 3, 6),
            'alto': max(0, min(1, (n_repeticiones - 3) / 5))
        }

    def fuzzificar_longitud(self, n_palabras):
        """Convierte longitud del texto a grados difusos"""
        return {
            'corto': max(0, 1 - n_palabras / 10),
            'medio': self.membresia_triangular(n_palabras, 5, 15, 30),
            'largo': max(0, min(1, (n_palabras - 20) / 40))
        }

    def fuzzificar_emojis(self, n_emojis, n_palabras):
        """Convierte densidad de emojis a grados difusos"""
        densidad = n_emojis / max(n_palabras, 1)
        return {
            'bajo': max(0, 1 - densidad * 5),
            'medio': self.membresia_triangular(densidad, 0.1, 0.3, 0.6),
            'alto': max(0, min(1, (densidad - 0.3) / 0.5))
        }

    # ==================== FUZZIFICACIÓN LINGÜÍSTICA ====================

    def fuzzificar_intensificadores(self, n_intensificadores):
        """Convierte número de intensificadores a grados difusos"""
        return {
            'ninguno': max(0, 1 - n_intensificadores),
            'pocos': self.membresia_triangular(n_intensificadores, 0, 1, 3),
            'muchos': max(0, min(1, (n_intensificadores - 1) / 3))
        }

    def fuzzificar_atenuadores(self, n_atenuadores):
        """Convierte número de atenuadores a grados difusos"""
        return {
            'ninguno': max(0, 1 - n_atenuadores),
            'pocos': self.membresia_triangular(n_atenuadores, 0, 1, 3),
            'muchos': max(0, min(1, (n_atenuadores - 1) / 3))
        }

    def fuzzificar_negadores(self, n_negadores):
        """Convierte número de negadores a grados difusos"""
        return {
            'ninguno': max(0, 1 - n_negadores),
            'presentes': max(0, min(1, n_negadores))
        }

    # ==================== 15 REGLAS DIFUSAS MEJORADAS ====================

    def aplicar_15_reglas_difusas(self, fuzzy_vars):
        """
        15 reglas difusas que combinan:
        - Características paralingüísticas
        - Modificadores lingüísticos
        """
        reglas_high = []
        reglas_medium = []
        reglas_low = []

        # ===== REGLAS HIGH (Intensidad Alta) =====

        # REGLA 1: Muchas exclamaciones + Mayúsculas altas
        reglas_high.append(min(
            fuzzy_vars['exclamaciones']['muchas'],
            fuzzy_vars['mayusculas']['alto']
        ))

        # REGLA 2: Repeticiones altas + Exclamaciones + Emojis altos
        reglas_high.append(min(
            fuzzy_vars['repeticiones']['alto'],
            fuzzy_vars['exclamaciones']['moderadas'],
            fuzzy_vars['emojis']['alto']
        ))

        # REGLA 3: Intensificadores + Exclamaciones (sin atenuadores)
        reglas_high.append(min(
            fuzzy_vars['intensificadores']['muchos'],
            fuzzy_vars['exclamaciones']['muchas'],
            fuzzy_vars['atenuadores']['ninguno']
        ))

        # REGLA 4: Texto corto + Exclamaciones + Mayúsculas
        reglas_high.append(min(
            fuzzy_vars['longitud']['corto'],
            fuzzy_vars['exclamaciones']['muchas'],
            fuzzy_vars['mayusculas']['alto']
        ))

        # REGLA 5: Muchos emojis + Exclamaciones (sin negadores)
        reglas_high.append(min(
            fuzzy_vars['emojis']['alto'],
            fuzzy_vars['exclamaciones']['moderadas'],
            fuzzy_vars['negadores']['ninguno']
        ))

        # ===== REGLAS MEDIUM (Intensidad Media) =====

        # REGLA 6: Exclamaciones moderadas + Mayúsculas medias
        reglas_medium.append(min(
            fuzzy_vars['exclamaciones']['moderadas'],
            fuzzy_vars['mayusculas']['medio']
        ))

        # REGLA 7: Intensificadores pocos + Exclamaciones pocas
        reglas_medium.append(min(
            fuzzy_vars['intensificadores']['pocos'],
            fuzzy_vars['exclamaciones']['pocas']
        ))

        # REGLA 8: Emojis medios + Texto medio
        reglas_medium.append(min(
            fuzzy_vars['emojis']['medio'],
            fuzzy_vars['longitud']['medio']
        ))

        # REGLA 9: Repeticiones medias OR (Exclamaciones pocas + Emojis medios)
        reglas_medium.append(max(
            fuzzy_vars['repeticiones']['medio'],
            min(fuzzy_vars['exclamaciones']['pocas'], fuzzy_vars['emojis']['medio'])
        ))

        # REGLA 10: Atenuadores presentes + Exclamaciones moderadas
        reglas_medium.append(min(
            fuzzy_vars['atenuadores']['pocos'],
            fuzzy_vars['exclamaciones']['moderadas']
        ))

        # ===== REGLAS LOW (Intensidad Baja) =====

        # REGLA 11: Sin exclamaciones + Sin mayúsculas
        reglas_low.append(min(
            fuzzy_vars['exclamaciones']['ninguna'],
            fuzzy_vars['mayusculas']['bajo']
        ))

        # REGLA 12: Texto largo + Pocas repeticiones + Pocas exclamaciones
        reglas_low.append(min(
            fuzzy_vars['longitud']['largo'],
            fuzzy_vars['repeticiones']['bajo'],
            fuzzy_vars['exclamaciones']['pocas']
        ))

        # REGLA 13: Atenuadores presentes + Sin intensificadores
        reglas_low.append(min(
            fuzzy_vars['atenuadores']['muchos'],
            fuzzy_vars['intensificadores']['ninguno']
        ))

        # REGLA 14: Negadores presentes + Pocos emojis
        reglas_low.append(min(
            fuzzy_vars['negadores']['presentes'],
            fuzzy_vars['emojis']['bajo']
        ))

        # REGLA 15: Texto medio + Sin puntuación enfática
        reglas_low.append(min(
            fuzzy_vars['longitud']['medio'],
            fuzzy_vars['exclamaciones']['ninguna'],
            fuzzy_vars['repeticiones']['bajo']
        ))

        # AGREGACIÓN (máximo de cada categoría)
        grado_high = max(reglas_high) if reglas_high else 0.0
        grado_medium = max(reglas_medium) if reglas_medium else 0.0
        grado_low = max(reglas_low) if reglas_low else 0.0

        return {
            'high': grado_high,
            'medium': grado_medium,
            'low': grado_low
        }

    # ==================== DEFUZZIFICACIÓN ====================

    def defuzzificar(self, grados_pertenencia):
        """Centro de Gravedad (CoG)"""
        valores = {'low': 1, 'medium': 2, 'high': 3}
        numerador = sum(grados_pertenencia[cat] * valores[cat] for cat in grados_pertenencia)
        denominador = sum(grados_pertenencia.values())
        return numerador / denominador if denominador > 0 else 1

    def corregir_texto(self, texto):
        return corregir_texto(texto, self.listas_correccion)


# =====================================================
# EXTRACTOR DE CARACTERÍSTICAS MEJORADO
# =====================================================

class ExtractorMejorado:
    """
    Extrae características paralingüísticas Y lingüísticas
    """

    def __init__(self):
        self.sistema_difuso = SistemaLogicaDifusaMejorado()

    def contar_modificadores(self, texto):
        """Cuenta intensificadores, atenuadores y negadores"""
        texto_lower = texto.lower()
        palabras = texto_lower.split()

        # Contar modificadores
        n_intensificadores = 0
        n_atenuadores = 0
        n_negadores = 0

        for idioma in ['en', 'es']:
            for palabra in palabras:
                if any(intensif in palabra for intensif in self.sistema_difuso.intensificadores[idioma]):
                    n_intensificadores += 1
                if any(atenu in palabra for atenu in self.sistema_difuso.atenuadores[idioma]):
                    n_atenuadores += 1
                if any(neg in palabra for neg in self.sistema_difuso.negadores[idioma]):
                    n_negadores += 1

        return n_intensificadores, n_atenuadores, n_negadores

    def extraer_caracteristicas(self, texto):
        """Extrae TODAS las características"""
        texto_lower = texto.lower()
        palabras = re.findall(r'\b\w+\b', texto_lower)

        # CARACTERÍSTICAS PARALINGÜÍSTICAS
        exclamaciones = texto.count('!')
        interrogaciones = texto.count('?')
        puntos_suspensivos = texto.count('...')

        # Mayúsculas
        palabras_mayusculas = sum(1 for p in palabras if p.isupper() and len(p) > 1)
        ratio_mayusculas = palabras_mayusculas / max(len(palabras), 1)

        # Repeticiones
        repeticiones_exclamacion = len(re.findall(r'!{2,}', texto))
        repeticiones_interrogacion = len(re.findall(r'\?{2,}', texto))
        letras_repetidas = len(re.findall(r'(.)\1{2,}', texto_lower))
        total_repeticiones = repeticiones_exclamacion + repeticiones_interrogacion + letras_repetidas

        # Emojis
        emojis = len(re.findall(r'[😀-🙏🌀-🗿🚀-🛿]', texto))
        emoticones = len(re.findall(r'[:;=]-?[\)\(DPO]', texto))
        total_emojis = emojis + emoticones

        # CARACTERÍSTICAS LINGÜÍSTICAS
        n_intensificadores, n_atenuadores, n_negadores = self.contar_modificadores(texto)

        # FUZZIFICACIÓN
        fuzzy_exclamaciones = self.sistema_difuso.fuzzificar_exclamaciones(exclamaciones)
        fuzzy_mayusculas = self.sistema_difuso.fuzzificar_mayusculas(ratio_mayusculas)
        fuzzy_repeticiones = self.sistema_difuso.fuzzificar_repeticiones(total_repeticiones)
        fuzzy_longitud = self.sistema_difuso.fuzzificar_longitud(len(palabras))
        fuzzy_emojis = self.sistema_difuso.fuzzificar_emojis(total_emojis, len(palabras))

        fuzzy_intensificadores = self.sistema_difuso.fuzzificar_intensificadores(n_intensificadores)
        fuzzy_atenuadores = self.sistema_difuso.fuzzificar_atenuadores(n_atenuadores)
        fuzzy_negadores = self.sistema_difuso.fuzzificar_negadores(n_negadores)

        fuzzy_vars = {
            'exclamaciones': fuzzy_exclamaciones,
            'mayusculas': fuzzy_mayusculas,
            'repeticiones': fuzzy_repeticiones,
            'longitud': fuzzy_longitud,
            'emojis': fuzzy_emojis,
            'intensificadores': fuzzy_intensificadores,
            'atenuadores': fuzzy_atenuadores,
            'negadores': fuzzy_negadores
        }

        grados_inferidos = self.sistema_difuso.aplicar_15_reglas_difusas(fuzzy_vars)

        # FEATURES PARA RANDOM FOREST
        features = {}

        # Salidas de las 15 reglas difusas
        features['fuzzy_high'] = grados_inferidos['high']
        features['fuzzy_medium'] = grados_inferidos['medium']
        features['fuzzy_low'] = grados_inferidos['low']
        features['fuzzy_score_defuzz'] = self.sistema_difuso.defuzzificar(grados_inferidos)

        # Características paralingüísticas
        features['exclamaciones'] = exclamaciones
        features['interrogaciones'] = interrogaciones
        features['puntos_suspensivos'] = puntos_suspensivos
        features['total_emojis'] = total_emojis
        features['emojis_graficos'] = emojis
        features['emoticones_texto'] = emoticones
        features['densidad_emojis'] = total_emojis / max(len(palabras), 1)

        features['palabras_mayusculas'] = palabras_mayusculas
        features['ratio_mayusculas'] = ratio_mayusculas

        features['repeticiones_exclamacion'] = repeticiones_exclamacion
        features['repeticiones_interrogacion'] = repeticiones_interrogacion
        features['letras_repetidas'] = letras_repetidas
        features['total_repeticiones'] = total_repeticiones

        # Características lingüísticas
        features['intensificadores'] = n_intensificadores
        features['atenuadores'] = n_atenuadores
        features['negadores'] = n_negadores
        features['balance_modificadores'] = n_intensificadores - n_atenuadores
        features['tiene_negacion'] = 1 if n_negadores > 0 else 0

        # Características de longitud
        features['longitud_texto'] = len(texto)
        features['num_palabras'] = len(palabras)

        palabra_counts = Counter(palabras)
        features['ratio_repeticion_palabras'] = 1 - (len(palabra_counts) / max(len(palabras), 1))

        features['balance_puntuacion'] = exclamaciones - interrogaciones
        features['densidad_mayusculas'] = palabras_mayusculas / max(len(texto), 1)

        # Interacciones
        features['exclamaciones_x_mayusculas'] = exclamaciones * ratio_mayusculas
        features['intensif_x_exclamaciones'] = n_intensificadores * exclamaciones
        features['atenu_x_negadores'] = n_atenuadores * n_negadores

        return features


# =====================================================
# CARGAR Y PREPARAR DATOS
# =====================================================
print("="*80)
print("🎯 SISTEMA MEJORADO: BALANCEO + 15 REGLAS DIFUSAS + ANÁLISIS LINGÜÍSTICO")
print("="*80)

print("\n📥 Cargando dataset...")
ds = load_dataset("stepp1/tweet_emotion_intensity")
df_data = pd.DataFrame(ds['train'])

print(f"\n✅ Dataset cargado: {len(df_data)} muestras")
print(f"\n📊 Distribución ORIGINAL:")
print(df_data['sentiment_intensity'].value_counts())
print(df_data['sentiment_intensity'].value_counts(normalize=True))

# Limpiar
df_data = df_data.dropna(subset=['tweet', 'sentiment_intensity'])
df_data = df_data[df_data['tweet'].astype(str).str.strip() != '']

# ============================================================
# 🧠 APLICAR JARO-WINKLER PARA CORREGIR PALABRAS DE EMOCIONES
# ============================================================

print("\n🔧 Corrigiendo tweets con Jaro-Winkler (ortografía emocional)...")

sistema = SistemaLogicaDifusaMejorado()

df_data['tweet'] = df_data['tweet'].apply(sistema.corregir_texto)

print("✅ Corrección completada.")
# ============================================================


# ============================================================
# 🚀 BALANCEAR EL DATASET (SIN ALTERAR VARIABLES)
# ============================================================
from sklearn.utils import resample

print("\n⚖️ Balanceando clases...")

min_count = df_data['sentiment_intensity'].value_counts().min()
balanced_frames = []

for label in df_data['sentiment_intensity'].unique():
    balanced_frames.append(
        df_data[df_data['sentiment_intensity'] == label].sample(
            n=min_count,
            replace=False
        )
    )

df_data = pd.concat(balanced_frames).sample(frac=1, random_state=42)

print("\n📊 Distribución BALANCEADA:")
print(df_data['sentiment_intensity'].value_counts())
# ============================================================

# Separar en Train y Test
df_train, df_test = train_test_split(
    df_data,
    test_size=0.2,
    random_state=42,
    stratify=df_data['sentiment_intensity']
)

print(f"\n✂️ Separación Train/Test:")
print(f"   • Train: {len(df_train)}")
print(f"   • Test:  {len(df_test)}")

print("\n📊 Distribución en TRAIN:")
print(df_train['sentiment_intensity'].value_counts())

print("\n📊 Distribución en TEST:")
print(df_test['sentiment_intensity'].value_counts())

# =====================================================
# EXTRAER CARACTERÍSTICAS
# =====================================================

print("\n🔍 Extrayendo características mejoradas...")
extractor = ExtractorMejorado()

caracteristicas_train = [extractor.extraer_caracteristicas(texto) for texto in df_train['tweet']]
caracteristicas_test = [extractor.extraer_caracteristicas(texto) for texto in df_test['tweet']]

df_features_train = pd.DataFrame(caracteristicas_train)
df_features_test = pd.DataFrame(caracteristicas_test)

print(f"✅ {len(df_features_train.columns)} características extraídas")

X_train = df_features_train.values
y_train = df_train['sentiment_intensity'].values

X_test = df_features_test.values
y_test = df_test['sentiment_intensity'].values

# =====================================================
# BALANCEO CON SMOTE
# =====================================================

print("\n⚖️ BALANCEO DE DATOS CON SMOTE...")
print(f"\n📊 Distribución ANTES del balanceo:")
print(pd.Series(y_train).value_counts())

# SMOTE + Tomek Links (elimina ruido)
smote_tomek = SMOTETomek(random_state=42, sampling_strategy='auto')
X_train_balanced, y_train_balanced = smote_tomek.fit_resample(X_train, y_train)

print(f"\n📊 Distribución DESPUÉS del balanceo:")
print(pd.Series(y_train_balanced).value_counts())
print(f"\n✅ Muestras balanceadas: {len(X_train_balanced)}")

# Escalar
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_balanced)
X_test_scaled = scaler.transform(X_test)

# =====================================================
# RANDOM FOREST
# =====================================================

print("\n" + "="*80)
print("🌲 ENTRENAMIENTO RANDOM FOREST (Datos Balanceados)")
print("="*80)

rf_model = RandomForestClassifier(
    n_estimators=500,
    max_depth=15,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features='sqrt',
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)

print("\n⏳ Entrenando modelo...")
rf_model.fit(X_train_scaled, y_train_balanced)

y_pred_train = rf_model.predict(X_train_scaled)
y_pred_test = rf_model.predict(X_test_scaled)

print("\n📈 Resultados en ENTRENAMIENTO:")
print(classification_report(y_train_balanced, y_pred_train, digits=3))

print("\n📈 Resultados en PRUEBA:")
print(classification_report(y_test, y_pred_test, digits=3))

cv_scores = cross_val_score(rf_model, X_train_scaled, y_train_balanced, cv=5)
print(f"\n🎯 Validación cruzada: {cv_scores.mean():.1%} ± {cv_scores.std():.1%}")

# =====================================================
# ANÁLISIS DE IMPORTANCIA
# =====================================================

feature_importance = pd.DataFrame({
    'caracteristica': df_features_train.columns,
    'importancia': rf_model.feature_importances_
}).sort_values('importancia', ascending=False)

print("\n🔝 Top 20 características:")
print(feature_importance.head(20).to_string(index=False))

# =====================================================
# VISUALIZACIONES
# =====================================================

# Matriz de confusión
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

cm_train = confusion_matrix(y_train_balanced, y_pred_train, labels=['low', 'medium', 'high'])
cm_test = confusion_matrix(y_test, y_pred_test, labels=['low', 'medium', 'high'])

sns.heatmap(cm_train, annot=True, fmt='d', ax=axes[0], cmap='Blues',
            xticklabels=['low', 'medium', 'high'],
            yticklabels=['low', 'medium', 'high'])
axes[0].set_title('ENTRENAMIENTO (Datos Balanceados)', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Real')
axes[0].set_xlabel('Predicha')

accuracy_test = accuracy_score(y_test, y_pred_test)
sns.heatmap(cm_test, annot=True, fmt='d', ax=axes[1], cmap='Greens',
            xticklabels=['low', 'medium', 'high'],
            yticklabels=['low', 'medium', 'high'])
axes[1].set_title(f'PRUEBA\nAccuracy: {accuracy_test:.1%}', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Real')
axes[1].set_xlabel('Predicha')

plt.tight_layout()
plt.savefig('matriz_confusion_mejorado.png', dpi=300, bbox_inches='tight')
print("\n✅ Guardado: matriz_confusion_mejorado.png")
plt.show()

# Importancia
plt.figure(figsize=(14, 12))
top_features = feature_importance.head(25)
fuzzy_features = ['fuzzy_high', 'fuzzy_medium', 'fuzzy_low', 'fuzzy_score_defuzz']
colors = ['#FF6B6B' if name in fuzzy_features else '#4ECDC4' for name in top_features['caracteristica']]

plt.barh(range(len(top_features)), top_features['importancia'].values, color=colors)
plt.yticks(range(len(top_features)), top_features['caracteristica'].values)
plt.xlabel('Importancia')
plt.title('Top 25 Características\n🔴 Lógica Difusa | 🔵 Paralingüísticas + Lingüísticas',
          fontsize=14, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('importancia_mejorado.png', dpi=300, bbox_inches='tight')
print("✅ Guardado: importancia_mejorado.png")
plt.show()

# =====================================================
# FUNCIÓN DE PREDICCIÓN
# =====================================================

def predecir_intensidad(texto):
    """Predice intensidad con análisis completo"""
    features = extractor.extraer_caracteristicas(texto)
    features_array = np.array([list(features.values())])
    features_scaled = scaler.transform(features_array)

    prediccion = rf_model.predict(features_scaled)[0]
    probabilidades = rf_model.predict_proba(features_scaled)[0]

    clases = rf_model.classes_
    prob_dict = {clase: prob for clase, prob in zip(clases, probabilidades)}

    return {
        'intensidad': prediccion,
        'probabilidades': prob_dict,
        'grados_difusos': {
            'high': features['fuzzy_high'],
            'medium': features['fuzzy_medium'],
            'low': features['fuzzy_low']
        },
        'modificadores': {
            'intensificadores': features['intensificadores'],
            'atenuadores': features['atenuadores'],
            'negadores': features['negadores']
        }
    }

# =====================================================
# EJEMPLOS
# =====================================================

print("\n" + "="*80)
print("📝 EJEMPLOS DE PREDICCIÓN")
print("="*80)

ejemplos = [
    "This is okay",
    "I ABSOLUTELY LOVE THIS!!!",
    "Kinda disappointing...",
    "NOT great at all",
    "AMAZING!!! Best ever!!!"
]

for ejemplo in ejemplos:
    resultado = predecir_intensidad(ejemplo)
    print(f"\n📌 '{ejemplo}'")
    print(f"   → {resultado['intensidad'].upper()}")
    prob_str = " | ".join([f"{k[0].upper()}={v:.1%}" for k, v in resultado['probabilidades'].items()])
    print(f"   → {prob_str}")
    print(f"   → Modificadores: I={resultado['modificadores']['intensificadores']} "
          f"A={resultado['modificadores']['atenuadores']} N={resultado['modificadores']['negadores']}")

print("\n" + "="*80)
print("✨ RESUMEN FINAL")
print("="*80)
print(f"✅ 15 reglas difusas (paralingüísticas + lingüísticas)")
print(f"✅ Datos balanceados con SMOTE+Tomek")
print(f"✅ {len(df_features_train.columns)} características totales")
print(f"✅ Accuracy en test: {accuracy_test:.1%}")
print(f"✅ Mejora respecto al 30% inicial: +{(accuracy_test-0.30)*100:.0f} puntos")
print("="*80)
```
